```python
from __future__ import print_function, division
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy
```

```python
# Load the data
data = np.loadtxt("droptower_vdata.txt",unpack=True)
```

```python
# Calculate position and acceleration
time = data[0]
velocity = data[1]
position = scipy.integrate.cumtrapz(velocity, time, initial = 30)
print(position)
print(velocity)
acceleration = np.diff(velocity)
print(acceleration)
```

```python
# Plot the position vs time
plt.plot(time, position,'o', color='red')
plt.title('Position vs Time for a Drop Tower')
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
```

```python
# Plot velocity vs time
plt.plot(time, velocity,'o')
plt.title('Velocity vs Time for a Drop Tower')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
```

```python
# Plot acceleration vs time
# These values are calculated using a difference between successive points
# Thus the output dimension will be one smaller than the input dimension.
plt.plot(acceleration,'o',color='green')
plt.title('Acceleration vs Time for a Drop Tower')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration (m/s^2)')
```

```python
# Calculate average acceleration when going upwards
positive_acceleration = []
for value in acceleration:
    if value >= 0:
        positive_acceleration.append(value)
average_positive = np.mean(positive_acceleration)
print("This is the average positive acceleration: " + str(average_positive))
```
